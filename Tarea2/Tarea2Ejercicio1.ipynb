**2.1** Let $C \in R^n$ be a convex set, with $x_1,...,x_k \in C$, and let $\theta_1, ... ,\theta_k \in R$ satisfy $\theta_i \geq 0$, $\theta_1 + ... + \theta_k = 1$. Show that $\theta_1x_1 + ... + \theta_k x_k \in C$ ( The  deﬁnition  of  convexity  is  that this  holds  for  $k  =  2$;  you  must  show  it  for  arbitrary  k) *Hint*. Use induction on $k$


**Solución:** Utilicemos inducción sobre $k$, entonces:

Tomamos el caso base como $k=2$, pero este ya se tiene que es convexo porque el caso $k = 2$ es la definición de conjunto convexo.

Ahora supongamos que se cumple para el caso $k = n$ es decir que dado un conjunto convexo $C \in R^n$ con $x_1,...,x_n \in C$ y sea $\theta_1,...,\theta_n \in R$ que cumplen que $\theta_i \geq 0$, $\theta_1+ ...+ \theta_n = 1$, se tiene que $\theta_1x_1 + ... + \theta_n x_n \in C$ (Hipótesis de inducción).

Ahora probemos el caso $k = n+1$, es decir dado un conjunto convexo $C \in R^n$ con $x_1,...,x_n, x_{n+1} \in C$ y sea $\theta_1,...,\theta_n, \theta_{n+1} \in R$ que cumplen que $\theta_i \geq 0$, $\theta_1+ ...+ \theta_n + \theta_{n+1} = 1$, tenemos que probar que $\theta_1x_1 + ... + \theta_n x_n + \theta_{n+1} x_{n+1} \in C$. Primero es claro que por lo menos va a haber un $\theta \neq 1$ , entonces escojamos sin perdida de generalidad $\theta_1 \neq 1$, luego podemos reescribir la ecuación como:

$$ \theta_1x_1 + ... + \theta_n x_n + \theta_{n+1} x_{n+1} = \theta_1x_1 + (1-\theta_1)(\rho_2x_2+ ... + \rho_{n+1}x_{n+1})$$

de manera que $\rho_i = \theta_i/(1-\theta_1)$, y $(1-\theta_1) = \theta_2 + ... + \theta_{n+1}$. Ahora es claro que $\rho_2 + ... + \rho_{n+1} = (1-\theta_1) / (1-\theta_1) = 1$, entonces por esto (y como desde $i = 2,...,n+1$ hay $n$ elementos) entonces tenemos que $\rho_2x_2+ ... + \rho_{n+1}x_{n+1} \in C$  por Hipótesis de induccion. Luego con $y = \rho_2x_2+ ... + \rho_{n+1}x_{n+1}$ nos queda la ecuacion:

$$\theta_1x_1 + (1-\theta_1)(\rho_2x_2+ ... + \rho_{n+1}x_{n+1}) = \theta_1x_1 +  (1-\theta_1)y$$

y como $x_1, y \in C$ y $\theta_1 + (1-\theta_1) = 1$ entonces por definición de conjunto convexo tenemos que $\theta_1x_1 +  (1-\theta_1)y \in C$ (que es lo mismo a $\theta_1x_1 + ... + \theta_n x_n + \theta_{n+1} x_{n+1} \in C$) por lo tanto queda demostrado.

**2.3** *Midpoint convexity*. A set $C$ is *midpoint convex* if whenever two points $a,b$ are in $C$, the average or midpoint $\frac{(a+b)}{2}$ is in C. Obviously a convex set is midpoint convex. It can be proved that under mild conditions midpoint convexity implies convexity. As a simple case, prove that if $C$ is closed and midpoint convex, then $C$ is convex.

**Solución:** Lo que tenemos que mostrar es dado $x,y \in C$ con $C$ siendo cerrado y midpoint convex, se tiene que $\theta x + (1-\theta)y \in C$ con $\theta \in [0,1]$ entonces para esto construyamos la siguiente sucesión $\{s_i\}$:

$$s_0 = [a,b]$$ y $$ \theta \in s_0$$

$$ s_{i+1}=   \left\{
\begin{array}{ll}
      [a, \frac{a+b}{2}] & \text{ si } \theta \in [a, \frac{a+b}{2}] \\
      [\frac{a+b}{2}, b] & \text{ si } \theta \in (\frac{a+b}{2}, b] \\
\end{array} 
\right.  $$

Teniendo en cuenta que la secuencia es convergente y como esta construida a partir de los puntos medios entonces $s_i x + (1-s_i)y \in C$ luego sacando el límite de esta secuencia tenemos que:

$$\lim_{i \rightarrow \infty} s_i x + (1-s_i)y = \theta x + (1-\theta)y $$

Ahora como $C$ es cerrado entonces contiene límites de la secuencia de sus elementos por lo tanto queda demostrado que $\theta x + (1-\theta)y$ con $\theta \in [0,1]$ luego $C$ es convexo.

**2.4** Show  that  the  convex  hull  of  a  set  $S$   is  the  intersection  of  all  convex  sets  that  contain  $S$. (The  same  method  can  be  used  to  show  that  the  conic,  or  affine,  or  linear  hull  of  a  set  S is  the  intersection  of  all  conic  sets,  or  aﬃne  sets,  or  subspaces  that  contain  S.)

**Solución:** Sea $C_s$ la intersección de todos los conjuntos convexos que contienen a $S$, es decir $C_s = \bigcap\{A | A \text{ es convexo }, S \subseteq A \}$ entonces tenemos que demostrar que:
$$  conv(S) = C_s $$

Para demostrar esto probemos que ambas contenencias $ conv(S) \subseteq C_s$ $ conv(S) \supseteq C_s$ se tienen:

$\subseteq$ 

Sea $x \in conv(S)$ por lo tanto $x$ se puede escribir como $\theta_1x_1 + ... + \theta_kx_k$ con $x_1,...,x_k \in S$ (combinación convexa). Ahora sea un $A $ algún conjunto convexo tal que $S \subseteq A $, entonces se tiene que $x_1,...,x_k \in A$ luego como A es convexo y $x$ es combinación convexa de $x_1,...,x_k$ entonces $x \in A$. Ahora como A es un conjunto arbitrario, entonces $x$ esta en cada conjunto convexo $A$ tal que $S \subseteq A $, en particular, $x$ debe pertenecer a la intersección de todos estos conjuntos, entonces $x \in C_s$

$\supseteq$ 

Sea $x \in C_s$, como $C_s$ es la intersección de todos los conjuntos convexos que contienen a $S$ y la envolvente convexa de S $conv(S)$ es un conjunto convexo que contiene a $S$ entonces $x \in conv(S)$


Como se tienen las dos contenencias entonces se tiene $ conv(S) = C_s $ y queda demostrado.

**2.5** What   is   the   distance   between   two   parallel   hyperplanes $\{x \in R^n | a^Tx=b_1\}$ y $\{ x \in R^n | a^Tx = b_2\}$

**Solución:** Sea un $x_1$ un punto en el hiperplano $\{x \in R^n | a^Tx=b_1\}$, ahora si consideramos la ecuacion de linea $L$ que pasa por el punto $x_1$ en dirección del vector normal $a$ tenemos que $L = x_1 + at$ donde $t \in R$. Ahora para hallar el punto de intersección con el otro hiperplano notemos que es un punto de $L$ que cumple la ecuación del hiperplano $\{ x \in R^n | a^T = b_2\}$, entonces:

$$a^T(x_1 + at) = b_2 \iff a^Tx_1 + a^Tat = b_2$$
$$\iff a^Tx_1 + ||a||^2t = b_2$$
$$\iff ||a||^2t = b_2 - a^Tx_1$$
$$\iff t = \frac{b_2 - a^Tx_1}{||a||^2}$$
$$\iff t = \frac{b_2 - b_1}{||a||^2}$$

luego el punto de intersección $x_2$ que esta en el hiperplano $\{ x \in R^n | a^Tx = b_2\}$ va a estar dado por $x_2 = x_1 + a(\frac{b_2 - b_1}{||a||^2})$ luego la distancia entre estos 2 puntos es:

$$||x_1 - x_2|| = ||x_1 - x_1 + a(\frac{b_2 - b_1}{||a||^2})||$$
$$ = || a(\frac{b_2 - b_1}{||a||^2}) ||$$
$$ = \frac{|b_2-b_1|}{||a||^2} ||a|| $$
$$ = \frac{|b_2-b_1|}{||a||}$$

**2.9** *Voronoi   sets   and   polyhedral   decomposition*. Let $x_0,...,x_K \in R^n$. Consider the set of points that are closer (in Euclidean norm) to $x_0$ than the other $x_i$, *i.e*,

$$ V = \{ x \in R^n | ||x - x_0||_2 \leq ||x - x_i||_2, i = 1,...,K\}$$

$V$ is called the *Voronoi region* arround $x_0$ with respect to $x_1, ..., x_K$ 

**(a)** Show that $V$ is a polyhedron. Express $V$ in the form $V = \{ x | Ax \preceq b\}$

**Solución:** Veamos que dado $x_0$ y un $x_i$, $x$ va a estar mas cercano a $x_0$ que a otro $x_i$ si: $||x - x_0||_2 \leq ||x - x_i||_2$ y esto va a ser equivalente a $(x - x_0)^T(x - x_0) \leq (x - x_i)^T (x - x_i)$ por la norma euclidiana. Luego si tomamos un solo componente del vector y aplicando la norma tenemos:

$$ (x - x_0)^2 \leq (x - x_i)^2  = x^2 - 2x x_0 + x_0^2 \leq x^2 - 2 x x_i + x_i^2$$

Ahora despejamos los terminos con $x$ a un lado de la desigualdad:

$$ \iff - 2x x_0 - 2 x x_i \leq x_i^2-x_0^2$$
$$ \iff  2(x_0 - x_i) x\leq x_i^2-x_0^2$$

Luego por la norma:

$$ \iff 2 (x_0 - x_i)^T x \leq x_i^Tx_i-x_0^Tx_0$$ 

luego con esto se puede definir un semiespacio y podemos expresar $V$ como $V = \{x | Ax \preceq b\}$ con A y b  como:

$$ A = \begin{bmatrix} x_1 - x_0 \\ \vdots  \\ x_K - x_0 \end{bmatrix}  \qquad b = \begin{bmatrix} x_1^Tx_1-x_0^Tx_0 \\ \vdots  \\ x_K^Tx_K-x_0^Tx_0 \end{bmatrix}$$

**(b)** Conversely,  given a polyhedron $P$  with nonempty interior,  show how to ﬁnd $x_0,...,x_K$ so that the polyhedron is the Voronoi region of $x_0$ with respect to $x_1,...,x_K$.

**Solución:** Supongamos $V = \{x | Ax \preceq b\}$ con $A \in R^{Kxn}$ y $b \in R^K$. Ahora elijamos un $x_0 \in  \{x | Ax \prec b\}$, con esto construimos $K$ puntos $x_i$ tomando el inverso de $x_0$ respecto a los hiperplanos $\{x | a_i^Tx = b_i\}$, es decir que escojemos $x_i$ de la forma $x_i = x_0 + \lambda a_i$ donde $\lambda$ es tal que la distancia de $x_i$ al hiperplano definido por $a_{i}^Tx = b_i$ es igual a la distancia de $x_0$ al hiperplano:

$$ b_i - a_i^Tx_0 = a_i^T x_i - b_i$$

Aqui reemplazamos $x_i =  x_0 + \lambda x a_i$ y despejemos $\lambda$, entonces:

$$ b_i - a_i^Tx_0 = a_i^T x_i - b_i = b_i - a_i^Tx_0 = a_i^T( x_0 + \lambda x a_i) - b_i$$ 

$$\iff b_i - a_i^Tx_0 = a_i^T x_0 + \lambda a_i^T a_i - b_i$$

$$\iff 2 b_i - 2 a_i^Tx_0 = \lambda a_i^T a_i $$

$$\iff \frac{2(b_i - a_i^Tx_0)}{a_i^T a_i} = \lambda$$

$$\iff \frac{2(b_i - a_i^Tx_0)}{||a_i||^2} = \lambda$$

Con esto despejemos en $x_i = x_0 + \lambda x a_i$ y con esto hallamos los $x_i$ que son de la forma:

$$x_i = x_0 +\frac{2(b_i - a_i^Tx_0)}{||a_i||^2} (a_i) $$

 **2.10** *Solution set of a quadratic inequality*. Let $C \subseteq R^n$ be the solution set of a quadratic inequality, 
 
 $$ C = \{ x \in R^n | x^TAx + b^Tx + c \leq 0\}$$,
 
with $A\in S^n, b\in R^n$, and $c \in R$. (Are the converses of these statements true?)

**(a)** Show that $C$ is convex if $A \succeq 0$

**Solución:** Vamos a utilizar el hecho de que un conjunto es convexo si y solo su intersección con una linea arbitraria $\{r + tv | t \in R\}$ es convexo 

Tenemos que 

$$ (r + tv)^T A(r + tv) + b^T(r + tv) + c = \alpha t^2 + \beta t + \gamma$$

donde

$$ \alpha=v^TAv, \qquad \beta= b^Tv + 2r^TAv, \qquad \gamma = c+b^Tr + r^T A r$$,

luego la intersección de C con la linea definida por $r$ y v es el conjunto $\{ r + tv | \alpha t^2 + \beta t + \gamma \leq 0\}$

El cual es convexo si $\alpha \geq 0$. esto es verdad para algún v, si $v^TAv \geq 0$ para todo $v$ osea que $A \succeq 0$.

El reverso de esta proposición no se tiene, por ejemplo tome $A = -1, b=0, c=-1$. Luego $A \nsucceq 0$, pero $C = R$ es convexo.

**(b)** Show that the intersection of C and the hyperplane defined by $g^Tx+ h = 0$ (where $g \neq 0$) is convex if $A + \lambda gg^T \succeq 0$ for some $\lambda \in R$ 

**Solución:** Sea $H = \{x | g^Tx+ h = 0\}$. Definamos $\alpha, \beta$ y $\gamma$ como en la parte (a), y tambien definamos:

$$ \rho = g^T, \qquad \epsilon= g^Tr + h$$

Sin pérdida de generalidad podemos asumir que $r \in H$ es decir que $\epsilon = 0$. La intersección de $C \cap H$ con la linea definida por $r$ y $v$ es:

$$\{r + tv| \alpha t^2 + \beta t + \gamma, \rho t = 0\}$$

Si $\rho = g^Tv \neq 0$ la intersección es el singleton $\{r\}$, si $\gamma \leq 0 $, o es vacío. En otro caso es un conjunto convexo. Si $\rho = g^Tv = 0$, entonces el conjunto se reduce a 

$$\{r + tv| \alpha t^2 \beta t + \gamma \leq 0\}$$

el cual es convexo si $\alpha \geq 0$. Por lo tanto $C \cap H$ es convexo si 

$$g^Tv = 0 \rightarrow v^TAv \geq 0$$

Esto es verdad si existe un $\lambda$ tal que $A + \lambda gg^T \succeq 0 $ luego lo anterior se cumple, ya que

$$v^TAv = v^T(A + \lambda gg^T)v \geq 0$$

para todo $v$ que satisfaga $g^Tv = 0$

El inverso tampoco es verdad.

**2.13** *Conic hull of outer products*. Consider the set of rank-k *outer products*, defined as $\{ XX^T | X \in R^{n x k}, \text{ rank} X = k\}$. Describe its conic hull in simple terms.

**Solución:** Tenemos que $ XX^T \preceq 0$ y notemos que $\text{ rank} X = \text{ rank}XX^T =k$, una combinación positiva de este tipo de matrices puede tener rango hasta $n$ pero nunca menor que $k$, entonces sea $A$ y $B$ matrices semipositivas definidas de rango $k$ supongamos un $u \in Ker(A+B)$ (donde $Ker(x)$ es el kernel de $x$), entonces:

$$(A+B)v = 0 \iff v^T(A+B)v = 0 \iff v^TAv+ v^TBv = 0$$

this implies,

$$ v^TAv=0 \iff Av = 0, v^TBv=0 \iff Bv= 0$$

por lo tanto todo vector en $Ker(A+B)$ tiene que estar en $Ker(A)$ y en $Ker(B)$. Esto implica que $dim(Ker(A+B))$ no puede ser mas grande que $dim(Ker(A))$ o $dim(Ker(B))$, por lo tanto una combinación de matrices positivas semidefinidas solo puede ganar rango.

consiguiente es que la envolvente cónica del conjunto de productos exteriores de rango k es el conjunto de matrices semidefinidas positivas de rango mayor o igual que k, incluida la matriz 0.

**2.14** *Expanded and restricted sets.* Let $S \subseteq R^n$, and let $||\cdot||$ be a norm on $R^n$

**(a)** For $a \geq 0$ we define $S_a$ as $\{x| \text{ dist}(x,S) \leq a\}$, where dist$(x,S) = inf_{y\in S}||x-y||$. We refer to $S_a$ as $S$ *expanded or extended* by a. Show that if $S$ is convex, then $S_a$ is convex.

**Solución:** Sean $x_1,x_2 \in S_a$ , veamos que $x_1 \theta + x_2(1-\theta) \in S_a$ , entonces hay que probar:

$$ \text{ dist}(x_1 \theta + x_2(1-\theta),S) \leq a $$

Utilicemos la definicion:

$$
\begin{align}
    \text{ dist}(x_1 \theta + x_2(1-\theta),S) &= inf_{y\in S}||x_1 \theta + x_2(1-\theta)-y|| & \\
    &=inf_{y\in S}||x_1 \theta + x_2 - x_2\theta)-y|| & \\
    &=inf_{y\in S}||x_1 \theta + x_2 - x_2\theta)-y + y\theta - y \theta|| & \text{sumamos: } 0 = y\theta - y \theta \\
    &=inf_{y\in S}||\theta(x_1-y) + (1-\theta)(x_2-y)|| & \text{agrupamos terminos} \\
    &\leq inf_{y\in S}||\theta(x_1-y)|| + inf_{y\in S}||(1-\theta)(x_2-y)|| & \text{Desigualdad Cauchy-Schwarz}\\
    &= \theta inf_{y\in S}||x_1-y|| + (1-\theta) inf_{y\in S}||(x_2-y)|| & \text{sacamos constantes}\\
    &= \theta a + (1-\theta) a & \text{Definicion de }S_a \\
    &\leq a & \text{ya que }\theta + (1-\theta) = 1
\end{align}
$$

**(b)** For $a \geq 0$ we define $S_{-a} = \{x | B(x,a) \subseteq S\}$, where $B(x,a)$ is the ball (in the norm $||\cdot||$), centered at x, with radius a. We refer to $S_{-a}$ as $S$ *shrunk or restricted* by a, since $S_{-a}$ consists of all points that are at least a distance $a$ from $R^n \backslash S$. Show that if $S$ is convex, then $S_{-a}$ is convex.

**Solución:** Suponga $x_1,x_2 \in S_{-a}$, entonces hay que probar que $\theta x_1 + (1-\theta) x_2 \in S_{-a}$. Veamos que con estos puntos tenemos que para todo $b$ tal que $||b|| \leq a$ se tiene

$$ x_1 + b \in S  \qquad  x_2 + b \in S$$

Luego con $0 \leq \theta \leq 1$ y con $||b|| \leq a$ 

$$ \theta x_1 + (1-\theta)x_2 + b = \theta x_1 + (1-\theta)x_2 + \theta b + (1-\theta) b$$
$$ = \theta (x_1+b) + (1-\theta) (x_2+b)$$

Por lo tanto se tiene que $\theta (x_1+b) + (1-\theta) (x_2+b) \in S$ ya que $ x_1 + b \in S$ y  $ x_2 + b \in S$, y por lo que $S$ es convexo. En consecuencia de esto se tiene que $\theta x_1 + (1-\theta) x_2 \in S_{-a}$ por como estan definidos estos puntos.

**2.15** *Some sets of probability distributions.* Let $x$ be a real-valued random variable with **prob**$(x=a_i) = p_i, i = 1,...,n$, where $a_1 < a_2 < ... < a_n$. Of course $p \in R^n$ lies in the standard probability simplex $P = \{p | 1^Tp = 1 , p \succeq 0 \}$. Which of the following conditions are convex in p? (That is, for which of the following conditions is the set of $p \in P$ that satisfy the condition convex?)

**(a)** $\alpha \leq \textbf{E}f(x) \leq \beta$ where **E**$f(x)$ is the expected value of $f(x)$ *i.e.* **E**$f(x) = \sum_{i=1}^n p_if(a_i).$ (The function $f: R \rightarrow R$ is given).

**Solución:** Tomando la condición $A = \{p\in P | \alpha \leq Ef(x) \leq \beta\}$, entonces tomemos $x,y \in A$ entonces hay que mirar que con $\theta \in [0,1]$ que $\theta x + (1-\theta) y \in A$, primero como $x,y \in A$ entonces tenemos que:

$$ \alpha \leq Ef(x) \leq \beta \qquad \text{y} \qquad \alpha \leq Ef(y) \leq \beta$$

Ahora multipliquemos por $\theta$ la primera desigualdad (con $x$) y la segunda (con $y$) la multiplicamos por $(1-\theta)$ asi:

$$ \theta \alpha \leq \theta Ef(x) \leq \theta \beta \qquad \text{y} \qquad (1-\theta) \alpha \leq (1-\theta) Ef(y) \leq (1-\theta) \beta$$

Luego aplicando la definición tenemos:

$$ \theta \alpha \leq \theta \sum_{i=1}^n x_if(a_i). \leq \theta \beta \qquad \text{y} \qquad (1-\theta) \alpha \leq (1-\theta) \sum_{i=1}^n x_if(a_i) \leq (1-\theta) \beta$$

Después sumando ambas desigualdades tenemos que:

$$ \theta \alpha + (1-\theta) \alpha \leq \theta \sum_{i=1}^n x_if(a_i) + (1-\theta) \sum_{i=1}^n y_if(a_i). \leq \theta \beta + (1-\theta) \beta$$

Como $\theta \in [0,1]$ se tiene que:

$$ \alpha \leq \theta \sum_{i=1}^n x_if(a_i) + (1-\theta) \sum_{i=1}^n y_if(a_i). \leq \beta $$

Por linealidad y agrupando terminos tenemos:

$$ \alpha \leq \sum_{i=1}^n (\theta x_i + (1-\theta) y_i) f(a_i) \leq \beta $$

Por consecuencia $\theta x + (1-\theta) y \in A $ Y por lo tanto la condición satisface la convexidad.

**(b)** **prob**$(x > \alpha) \leq \beta$.

**Solución:** Veamos que con la condición $A = \{p\in p | prob(x > \alpha) \leq \beta \}$ va a suceder algo parecido que en (a) entonces sean $x,y \in A$ tenemos 2 desigualdades que por la definición de prob:

$$ \sum_{a_i > a} x_i \leq \beta  \qquad \text{y} \qquad \sum_{a_i > a} y_i \leq \beta $$

Luego multiplicando $\theta$ en la primera desigualdad y $(1-\theta)$ en la segunda desigualdad, tenemos: 

$$ \theta \sum_{a_i > a} x_i \leq \theta \beta  \qquad \text{y} \qquad (1-\theta) \sum_{a_i > a} y_i \leq (1-\theta) \beta $$

Sumamos ambas desigualdades luego:

$$\theta \sum_{a_i > a}  x_i + (1-\theta) \sum_{a_i > a} y_i \leq \theta \beta + (1-\theta) \beta$$

Como $\theta \in [0,1]$ se tiene que:

$$\theta \sum_{a_i > a}  x_i + (1-\theta) \sum_{a_i > a} y_i \leq \beta$$

Por linealidad y agrupando terminos tenemos:

$$ \sum_{a_i > a}\theta x_i + (1-\theta) y_i $$

Por consecuencia $\theta x + (1-\theta) y \in A $ Y por lo tanto la condición satisface la convexidad.


**(c)** **E**$|x^3| \leq \alpha$ **E** $|x|$.

**Solución:** Veamos que con la condición $A = \{p\in p | E|x^3| \leq \alpha E|x| \}$ va a suceder algo parecido que en los subpuntos anteriores entonces sean $x,y \in A$ tenemos 2 desigualdades que por la definición de A:

$$ \sum_{i=1}^n x_i |a_i^3| \leq \alpha \sum_{i=1}^n x_i|a_i|  \qquad \text{y} \qquad \sum_{i=1}^n y_i |a_i^3| \leq \alpha \sum_{i=1}^n y_i|a_i| $$

Ahora multipliquemos por $\theta$ la primera desigualdad (con $x$) y la segunda (con $y$) la multiplicamos por $(1-\theta)$ asi:

$$ \theta \sum_{i=1}^n x_i |a_i^3| \leq \theta \alpha \sum_{i=1}^n x_i|a_i|  \qquad \text{y} \qquad (1-\theta) \sum_{i=1}^n y_i |a_i^3| \leq (1-\theta) \alpha \sum_{i=1}^n y_i|a_i| $$

Sumamos ambas desigualdades:

$$ \theta \sum_{i=1}^n x_i |a_i^3| + (1-\theta) \sum_{i=1}^n y_i |a_i^3| \leq \theta \alpha \sum_{i=1}^n x_i|a_i| + (1-\theta) \alpha \sum_{i=1}^n y_i|a_i| $$

Por linealidad tenemos:
 
$$ \sum_{i=1}^n (\theta) x_i |a_i^3| + (1-\theta)  y_i |a_i^3| \leq \alpha \sum_{i=1}^n (\theta ) x_i|a_i| + (1-\theta) y_i|a_i| $$ 

Agrupando:

$$ \sum_{i=1}^n ((\theta) x_i + (1-\theta)  y_i)|a_i^3| \leq \alpha \sum_{i=1}^n ((\theta ) x_i + (1-\theta) y_i)|a_i|   $$

Por consecuencia $\theta x + (1-\theta) y \in A $ Y por lo tanto la condición satisface la convexidad.


**(d)** **E**$x^2 \leq \alpha$.

**Solución:** Veamos que con la condición $A = \{p\in p | E x^2 \leq \alpha \}$ va a suceder algo parecido que en los subpuntos anteriores entonces sean $x,y \in A$ tenemos 2 desigualdades que por la definición de A:

$$ \sum_{i=1}^n x_i a_i^2 \leq \alpha  \qquad \text{y} \qquad \sum_{i=1}^n y_i a_i^2 \leq \alpha $$

Ahora multipliquemos por $\theta$ la primera desigualdad (con $x$) y la segunda (con $y$) la multiplicamos por $(1-\theta)$ asi:

$$ \theta \sum_{i=1}^n x_i a_i^2 \leq \theta \alpha \qquad \text{y} \qquad (1-\theta) \sum_{i=1}^n y_i a_i^2 \leq (1-\theta) \alpha $$

Sumamos ambas desigualdades:

$$ \theta \sum_{i=1}^n x_i a_i^2 + (1-\theta) \sum_{i=1}^n y_i a_i^2 \leq \theta \alpha + (1-\theta) \alpha $$

Por linealidad tenemos y como $\theta \in [0,1]$ se tiene que:
 
$$ \sum_{i=1}^n (\theta) x_i a_i^2 + (1-\theta)  y_i a_i^2 \leq \alpha $$ 

Agrupando:

$$ \sum_{i=1}^n ((\theta) x_i + (1-\theta)  y_i)a_i^2 \leq \alpha $$

Por consecuencia $\theta x + (1-\theta) y \in A $ Y por lo tanto la condición satisface la convexidad.


**(e)** **E**$x^2 \geq \alpha$.

**Solución:** Veamos que con la condición $A = \{p\in p | E x^2 \leq \alpha \}$ va a suceder algo parecido que en los subpuntos anteriores entonces sean $x,y \in A$ tenemos 2 desigualdades que por la definición de A:

$$ \sum_{i=1}^n x_i a_i^2 \geq \alpha  \qquad \text{y} \qquad \sum_{i=1}^n y_i a_i^2 \geq \alpha $$

Ahora multipliquemos por $\theta$ la primera desigualdad (con $x$) y la segunda (con $y$) la multiplicamos por $(1-\theta)$ asi:

$$ \theta \sum_{i=1}^n x_i a_i^2 \geq \theta \alpha \qquad \text{y} \qquad (1-\theta) \sum_{i=1}^n y_i a_i^2 \geq (1-\theta) \alpha $$

Sumamos ambas desigualdades:

$$ \theta \sum_{i=1}^n x_i a_i^2 + (1-\theta) \sum_{i=1}^n y_i a_i^2 \geq \theta \alpha + (1-\theta) \alpha $$

Por linealidad tenemos y como $\theta \in [0,1]$ se tiene que:
 
$$ \sum_{i=1}^n (\theta) x_i a_i^2 + (1-\theta)  y_i a_i^2 \geq \alpha $$ 

Agrupando:

$$ \sum_{i=1}^n ((\theta) x_i + (1-\theta)  y_i)a_i^2 \geq \alpha $$

Por consecuencia $\theta x + (1-\theta) y \in A $ Y por lo tanto la condición satisface la convexidad.

**(f)** **var**$(x) \leq \alpha$, where **var**$(x) =$ **E**$(x - \textbf{E}x)^2$ is the variance of x.

**Solución:** Primero notemos que la restricción de la varianza va a estar dada por:

$$var(x) = Ex^2 - (Ex)^2 = \sum_{i=1}^n p_ia_i^2 - ( \sum_{i=1}^n p_i a_i^2)^2 \leq \alpha$$

Notemos que esto no es convexo en general, como un contraejemplo tomemos $n=2, a_1 = 0, a_2 = 1$ y $\alpha = 1/5$. $p = (1,0)$ y $p = (0,1)$ son dos puntos que satisfacen $var(x) \leq \alpha$ pero la combinación convexa $p = (1/2, 1/2)$ no lo cumple

**(g)** **var**$(x) \geq \alpha$ 

**Solución:** Esta restricción es equivalente a 

$$\sum_{i=1}^n a_i^2p_i + (\sum_{i=1}^n a_i p_i)^2 = b^Tp + + p^TAp \leq \alpha$$

donde $b_i = a_i^2$ y $A = aa^T$. Esto define un conjunto convexo ya que la matriz $aa^T$ es positiva semidefinida

**(h)** **quartile**$(x) \geq \alpha$, where **quartile**$(x) = inf\{ \beta | \textbf{prob}(x\leq \beta) \geq 0.25\}$

**Solución:** Notemos que la restricción $f(p) \geq \alpha$ es equivalente a $prob(x \leq \beta) < 0.25 \text{ para todo } \beta < \alpha$

Si $\alpha \geq \alpha_1$ es siempre verdad. ahora si no se cumple esto entonces definimos $k = max\{i | \alpha_i \leq \alpha\} $. Esto va a ser un entero independiente de $p$. Luego la condición $f(p) \geq \alpha$ se cumple si y solo si

$$prob(x \leq \alpha_k) = \sum_{i=1}^k p_i < 0.25$$

Esto va a ser una desigualdad lineal en p, y por lo tanto va a definir un semiespacio, y va a ser convexo

**(i)** **quartile**$(x) \leq \alpha$

**Solución:** Notemos que la restricción $f(p) \geq \alpha$ es equivalente a $prob(x \leq \beta) \geq 0.25 \text{ para todo } \beta \geq \alpha$

Luego entonces podemos expresar la restricción de la siguiente manera:

$$prob(x \geq \alpha_k) = \sum_{i=1}^k p_i \geq 0.25$$

Esto va a ser una desigualdad lineal en p, y por lo tanto va a definir un semiespacio, y va a ser convexo (y si $\alpha \leq \alpha_1$, definimos $k=0$)

**2.16** Show that if $S_1$ and $S_2$ are convex sets in $R^{mxn}$, then so is their partial sum

$$ S = \{ (x, y_1+y_2) | x \in R^m, y_1, y_2 \in R^n, (x,y_1) \in S_1, (x,y_2) \in S_2\}$$

**Solución:** Consideremos 2 puntos: $(a,b_1 + b_2), (c, d_1 + d_2) \in S$ tales que:

$$(a,b_1) \in S_1 \quad (a,b_2) \in S_2 \quad (c,d_1) \in S_1 \quad (c,d_2) \in S_2$$

Entonces para $\theta \in [0,1]$ tenemos que probar que $\theta(a,b_1+b_2) + (1-\theta)(c,d_1+d_2) \in S$, entonces realizando operaciones por componentes:

$$\theta(a,b_1+b_2) + (1-\theta)(c,d_1+d_2) = (\theta a + (1-\theta) c, (\theta b_1 + (1-\theta) d_1) + (\theta b_2 + (1-\theta) d_2))$$

Ahora como $(a,b_1) \in S_1$ y $ (c,d_1) \in S_1$ entonces $(\theta a + (1-\theta) c, \theta b_1 + (1-\theta) d_1) \in S_1$ por convexidad de $S_1$ y como $ (a,b_2) \in S_2 $ y $ (c,d_2) \in S_2 $ entonces $(\theta a + (1-\theta) c, \theta b_2 + (1-\theta) d_2) \in S_2$ por convexidad de $S_2$ 

Por lo tanto se va a tener que $\theta(a,b_1+b_2) + (1-\theta)(c,d_1+d_2) \in S$ entonces se cumple la convexidad.

**2.17** *Image of polyhedral sets under perspective function*. In this problem we study the image of hyperplanes, halfspaces and polyhedra under the perspective function $P(x,t) = x/t$, with **dom**$P = R^n x R_{++}$. For each of the following sets $C$, give a simple description of

$$P(C) = \{v/t | (v,t) \in C, t > 0\}$$

**(a)** The polyhedron $C = \textbf{conv}\{(v_1,t_1),....,(v_K,t_K)\}$ where $v_i \in R^n$ and $t_i > 0$.

**Solución:**  Veamos que podemos definir $P$ dividiendo por $t$ en $v$ por lo tanto queda de la siguiente manera: 

$$P(C) = conv\{v_1/t_1, v_2/t_2, ... ,v_K/t_K\}$$

Y esto se puede hacer gracias a la linearidad de la envolvente convexa por lo que esta definida asi:

$$conv\{v_1/t_1, v_2/t_2, ... ,v_K/t_K\} = \{ \sum_{i=1}^{K}\theta_i v_K/t_K | \theta_i \geq 0, 1^T\theta = 1 \}$$

**(b)** The hyperplane $C = \{(v,t) | f^Tv + gt = h\}$ (with $f$ and $g$ not both zero).

**Solución:** Veamos que podemos definir $P$ dividiendo por $t$ en $f^Tv + gt = h$ por lo tanto queda de la siguiente manera:

$$ f^Tv/t + gt/t = h/t \iff f^T z + g = h/t \quad \text{con $z = v/t$ }$$

Entonces $P(C) = \{z | f^Tz+ g = h/t \text{ para algún } t>0\}$
Y esto es equivalente a 

$$P(C) = \begin{cases}
  \{z|f^T+g = 0\}  & h=0  \\
  \{z|f^T+g > 0\} & h > 0\\
  \{z|f^T+g < 0\} & h < 0
\end{cases}$$

**(c)** The halfspace $C = \{(v,t) | f^Tv+ gt \leq h\}$ (with $f$ and $g$ not both zero).

**Solución:** Veamos que podemos definir $P$ dividiendo por $t$ en $f^Tv + gt \leq h$ por lo tanto queda de la siguiente manera:

$$ f^Tv/t + gt/t \leq h/t \iff f^T z + g \leq h/t \quad \text{con $z = v/t$ }$$

Entonces $P(C) = \{z | f^Tz+ g \leq h/t \text{ para algún } t>0\}$
Y esto es equivalente a 

$$P(C) = \begin{cases}
  \{z|f^T+g \leq 0\}  & h=0  \\
  R^n & h > 0\\
  \{z|f^T+g < 0\} & h < 0
\end{cases}$$

**(d)** The polyhedron $C = \{(v,t) | Fv + gt \preceq h\}$

**Solución:** Veamos que en este caso definimos $P(C)$ como:

$$ P(C) = \{ z | Fz + g \preceq (1/t)h \text{ para algún } t>0\} $$

Más explicitamente, $p \in P(C)$ si y solo si satisface las siguientes condiciones

1. $ f_i^T z + g_i \leq 0 $ si $h_i = 0$
2. $ f_i^T z + g_i < 0$ si $h_i < 0$
3. $ (f_i^T z + g_i) / h_i \leq (f_k^T z + g_k)/h_k $ si $h_i > 0$ y $h_k < 0$

**2.18** *Invertible linear-fractional functions.* Let $f: R^n \rightarrow R^n$ be the linear-fractional function 

$$f(x) = (Ax+b)/(c^Tx+d) \qquad \textbf{dom}f=\{x | c^Tx+d > 0\}$$

Suppose the matrix 

$$ Q = \begin{bmatrix} A & b \\ C^T & d  \end{bmatrix}$$

is nonsingular. Show that $f$ is invertible and that $f^{-1}$ is a linear-fractional mapping. Give an explicit expression for $f^{-1}$ and its domain in terms of $A,b,c$ and $d$. *Hint*. It may be easier to express $f^{-1}$ in terms of $Q$

**Solución:** Veamos que podemos definir $f$ con la matriz $Q$ con esto tenemos:

$$f(x) = P^{-1}(QP(x))$$

donde $P$ esta definido como la correspondencia proyectiva entre $R^n$ y el semiespacio de rayos con ultimo componente positivo, $P(z) = \{t(z,1) | t>0\}$. Luego con esto es posible ver que la inversa va a estar dada por (y gracias a que Q es no singular):

$$f^{-1}(x) = P^{-1}(Q^{-1}P(X)) $$
