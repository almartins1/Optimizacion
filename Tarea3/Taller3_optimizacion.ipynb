**3.4** Show that a continuous function $f: R^n \leftarrow R$ is convex if and only if for every line segment, its average value on the segment is less than or equal to the average of its values at the endpoints of the segment: For every $x, y \in R^n$.

$$ \int_0^1 f(x+\lambda(y-x)) d\lambda \leq \frac{f(x) + f(y)}{2}$$

**Solucion:**

$\Rightarrow)$ Supongamos que la función $f$ es convexa, por la desigualdad de Jensen tenemos que

$f(\lambda y + (1- \lambda)x)=f(x+\lambda(y-x))\leq \lambda f(y) + (1- \lambda)f(x)=f(x)+ \lambda (f(y)-f(x))$ para $0\leq \lambda \leq 1$.

Ahora, al integrar de $0$ a $1$ a ambas partes de la desigualdad obtenemos:

$\int_0^1 f(x+\lambda(y-x)) d \lambda \leq \int_0^1 (f(x)+ \lambda (f(y)-f(x)))d \lambda = \dfrac{f(y)+f(x)}{2}$.

$⇐)$ (Por contrareciproca) Supongamos que $f$ no es convexa y veamos que el promedio de $f$ sobre el segmento es mayor que el promedio de sus valores en los extremos:

Como $f$ no es convexa, existen $x,y, \gamma \in (0,1)$ tal que $f(\gamma x + (1- \gamma)y)>\gamma f(x) + (1- \lambda ) f(y)$.

Consideremos la función de $\theta$ dada por:

$F(\theta)=f(\theta x+(1-\theta)y)-\theta f(x)-(1-\theta)f(y)$, note que $F$ es continua pues $f$ es continua. Ademas, $F$ se hace ceroe en $\theta=0$ y $\theta=1$, y $F(\theta)>0$ cuando $\theta=\gamma$.

Sean $\alpha>\gamma$ el mayor cero de $F$ y $b<\gamma$ el menor cero de $F$, definamos asi $u=\alpha x + (1-\alpha )y$ y $v=\beta x + (1- \beta) y$. Sobre $(\alpha, \beta)$ tenemos

$F(θ) = f(θx + (1 − θ)y) > θf(x) + (1 − θ)f(y)$, de modo que para $θ ∈ (0, 1)$ obtenemos 

$f(θu + (1 − θ)v) > θf(u) + (1 − θ)f(v)$. Finalmente al integrar esta desigualdad desde $\theta=0$ hasta $\theta=1$:

$\int_0^1 f(u + \theta (u-v))d\theta>\int_0^1(f(u)+\theta(f(u)-f(v)))d\theta=\dfrac{f(u)+f(v)}{2}$

**3.6** *Functions and epigraphs.* When is the epigraph of a function a halfspace? When is the epigraph of a function of a convex cone? When is the epigraph of a function a polyhedron?

**Solución:**

a. El epigrafo de una función $f$ es un semiespacio si y solo si $f$ es una función a fin:

$⇒)$ Supongamos que $epi(f)$ es un semiespacio, esto nos dice que 

$(x,t)\in epi(f) \iff a^T\textbf{x}\leq t-b$, para algún $a \in R^n$, $a$ distinto de $0$ y $b \in R$ $\iff$ $a^T+b \leq t \iff f(x) \leq t$. 

$⇐)$ Supongamos que $f$ es una función a fin y veamos que su epigrafo es un semiespacio:

$(x,t)\in epi(f) \iff f(x) \leq t \iff c^T+d \leq t$ , para algún $c \in R^n$,  y $d \in R$ $\iff$ $c^T\textbf{x}\leq t-d$ . 

b. El epigrafo de una función $f$ es un cono convexo si y solo si la función es convexa y positivamente homogenea, es decir, $f(\alpha x)=\alpha f(x) ∀\alpha\geq0$:

Primero, recordemos que una función es convexa si y solo si su epigrafo es convexo.

$⇒)$ Supongamos que el epigrafo de $f$ es un cono, asi $∀ \alpha > 0$ y si $(x,t) \in epi(f)$ tenemos que $(\alpha x, \alpha t) \in epi(f)$. Luego, como $(x,f(x)) \in epi(f)$, entonces $(\alpha x, \alpha f(x)) \in epi(f)$ de modo que $f(\alpha x) \leq \alpha f(x)$ (1).

Ademas, como $(\alpha x, \alpha f(x)) \in epi(f)$ y $epi(f)$ es un cono entonces $(x, f(\alpha x)/\alpha) \in epi(f)$, es decir, $f(x)\leq \dfrac{f(\alpha x)}{\alpha}$ (2).

Luego, por (1) y (2) obtenemos $f(x) \leq \dfrac{f(\alpha x)}{\alpha} \leq f(x)$. Es decir, $f(\alpha x)= \alpha f(x)$.

$⇐)$ Supongamos que $f$ es una función positivamente homogenea y veamos que su epigrafo es un cono:

$(x,t) \in epi(f) \iff f(x) \leq t \iff \alpha f(x) = \alpha f(x) \leq \alpha t \iff (\alpha x, \alpha t) \in epi(f) ∀ \alpha >0$.

c. El epigrafo de una función $f$ es un poliedro si y solo si $f$ es una función afin a trozos:

Recordemos que la función $f(x)=\max \{a_1^Tx+b_1,\cdots, a_L^Tx+b_L\}$ define una función afin a trozos, con $L$ o menos regiones.

**3.13** *Kullback-Leibler divergence and the information inequality*. Let $D_{kl}$ be the Kullback-Leibler divergence, as defined in (3.17). Prove the *information inequality*: $D_{kl}(u,v) \geq 0$ for all $u,v \in R_{++}^n.$ Also show that $D_{kl}(u,v) = 0$ if and only if $u=v$.

*Hint*. The Kullback-Leibler divergence can be expressed as:

$$ D_{kl}(u,v) = f(u) - f(v) - \triangledown f(v)^T(u-v),$$

where $f(v) = \sum_{i=1}^n v_i log (v_i)$ is the negative entropy of $v$.

**Solución**:

Tengamos en cuenta que la definición de la divergencia de Kullback-Leibler es:

$$ D_{kl}(u,v) = \sum_{i=1}^n (u_i log(u_i/v_i) -u_i + v_i$$

Teniendo en cuenta que la entropia negativa $f(v) = \sum_{i=1}^n v_ilogv_i $ es convexa (ya que si tomamos la función $f(x) = xlogx$ y aplicamos el criterio del hessiano tenemos que $f''(x) = 1/x$  y esto siempre va a ser positivo para cualquier x positivo), utilicemos el critero de primer orden para convexidad, luego:

$$ f(u) > f(v) + \nabla f(v)^T(u-v)$$

para cualquier $u,v \in R_{++}^n con $u \neq v$. Luego evaluando en ambos lados tenemos:

$$ \begin{align*}
\sum_{i=1}^n u_ilog (u_i) &> \sum_{i=1}^n v_i + \sum_{i=1}^n (log (v_i) + 1)(u_i - v_i)\\
&= \sum_{i=1}^n (u_i log (v_i) + u_i - v_i)\\
\end{align*}$$

Luego multiplicando por -1 y pasando $\sum_{i=1}^n u_ilog (u_i)$ al otro lado de la desigualdad tenemos:

$$\begin{align*}
0 &< \sum_{i=1}^n u_ilog (u_i) - u_i log (v_i) -u_i + v_i\\
&= \sum_{i=1}^n (u_i log(u_i/v_i) -u_i + v_i\\
&= D_{kl}(u,v)
\end{align*}$$

Por lo tanto se tiene que $D_{kl}(u,v) > 0$ para $u \neq v$. Ahora si reemplazamos en $D_{kl}(u,v)$ con $u = v$ es claro que:

$$ D_{kl}(u,u) = \sum_{i=1}^n (u_i log(u_i/u_i) -u_i + u_i = \sum_{i=1}^n u_i log(1) - 0  = 0$$

**3.16** For each of the following functions determine whether it is convex, concave, quasiconvex, or quasiconcave.

(a) $f(x) = e^x - 1$ on $R$. <br>
(b) $f(x_1,x_2) = x_1x_2$ on $R{++}^2$ <br>
(c) $f(x_1,x_2) = 1/(x_1x_2)$ on $R{++}^2$ <br>
(d) $f(x_1,x_2) = x_1/x_2$ on $R{++}^2$ <br>
(e) $f(x_1,x_2) = x_1^2/x_2$ on $R x R_{++}$ <br>
(f) $f(x_1,x_2) = x_1^{\alpha}x_2^{1-\alpha},$ where $0 \leq 1$ on $R{++}^2$

**Solución:**

**(a)** Es claro que esta función es convexa ya que dado el criterio de segundo orden (Hessiano) tenemos:

$$f(x) = e^x-1 \qquad \rightarrow \qquad f'(x) = e^x \qquad \rightarrow f''(x) = e^x$$

y como $e^x \succeq 0$ entonces es definida positiva, por lo tanto es convexo. Como la función es convexa, también es quasiconvexa. Ahora veamos que la función es quasiconcava, entonces veamos las curvas de nivel de $-f$:

$$\begin{align*}
    S_\alpha &= \{x \in R | -e^x + 1 \leq \alpha\}\\
    &= \{x \in R | 1 - \alpha \leq e^x\}\\
    &= \{x \in R | ln(1-\alpha) \leq x\}
\end{align*}$$

Luego esto ultimo lo podemos expresar como el intervalo $[ln(1-\alpha) \leq x]$, y como es un intervalo entonces es convexo. Como las curvas de nivel de $-f$ son convexas entonces $f$ es quasiconcavo.
<br>
**(b)** Utilicemos otra vez el criterio de segundo orden, entonces con:
$$f(x,y) = xy \qquad dx = y \qquad dy = x$$

Luego el hessiano va a estar dado por:

$$\begin{bmatrix}
d^2x & dxdy \\
dydx & d^2y 
\end{bmatrix} = \begin{bmatrix}
0 & 1\\
1 & 0 
\end{bmatrix}$$

Pero es claro que no es definida positiva ya que no todas los determinantes de las submatrices principales son positivos($det\begin{bmatrix}
0 & 1\\
1 & 0 
\end{bmatrix} = 0-1 = -1$), luego esta función no es convexa. Similarmente tomando $-f$ y utilizando el mismo criterio tenemos que el hessiano va a estar dado por:
$$\begin{bmatrix}
d^2x & dxdy \\
dydx & d^2y 
\end{bmatrix} = \begin{bmatrix}
0 & -1\\
-1 & 0 
\end{bmatrix}$$

Pero similarmente se puede ver que esta matriz no va a ser definida positiva por lo tanto no es cóncavo.<br> 

Utilicemos criterio de primer orden (para quasiconcava) para verificar que $f$ es quasiconcavo, entonces tomando $-f$ tenemos que probar que dado
$$ f(y) \geq f(x) \rightarrow \nabla f(x)^T(y-x) \geq 0 $$
con esto:
$$\begin{align*}
\nabla f(x)^T(y-x) &= \begin{bmatrix}
-x_2 & -x_1 
\end{bmatrix} \begin{bmatrix}
y_1 - x_1 \\
y_2 - x_2 
\end{bmatrix}\\
&= 2x_1x_2 - x_2y_1 - x_1y_2
&= x_1x_2 (2 - \frac{y_1}{x_1}- \frac{y_2}{x_2})\\
&\leq x_1x_2 \left(2 - 2\sqrt{\frac{y_1y_2}{x_1x_2}} \right)
\end{align*}$$

Luego como tenemos que $f(y) \geq f(x) \rightarrow \frac{y_1y_2}{x_1x_2} \geq 1$ se tiene:

$$x_1x_2 \left(2 - 2\sqrt{\frac{y_1y_2}{x_1x_2}} \right) \leq x_1x_2(2-2) = 0$$

Entonces $f$ es quasiconcava. La función no va a ser quasiconvexa para ver esto podemos tomar la condición de primer orden (para quasiconvexa) y reemplazar algunos valores: Ej con $(y_1,y_2) = (4,3)$ y $(x_1,x_2) = (6,2)$ luego tenemos que:

$$2x_1x_2 - x_2y_1 - x_1y_2 = 2 > 0$$

**(c)** Utilicemos el criterio de segundo orden, entonces con:
$$f(x,y) = \frac{1}{xy} \qquad dx=\frac{-1}{x^2y} \qquad dy=\frac{-1}{xy^2}$$

Luego su hessiano va a estar dado por:

$$\begin{bmatrix}
d^2x & dxdy \\
dydx & d^2y 
\end{bmatrix} = \begin{bmatrix}
\frac{2}{x^3y} & \frac{1}{x^2y^2}\\
\frac{1}{x^2y^2} & \frac{2}{xy^3} 
\end{bmatrix}$$

Esta matriz es definida positiva ya que las determinantes de las submatrices principales superiores son positivas ($\frac{2}{x^3y}>0$ y $\frac{4}{x^4y^4} - \frac{1}{x^4y^4} = 3 >0$), luego la función es convexa y quasiconvexa.<br>

**(d)** Utilicemos el criterio de segundo orden, entonces con:
$$f(x,y) = \frac{x}{y} \qquad dx=\frac{1}{y} \qquad dy=\frac{-x}{y^2}$$

Luego su hessiano va a estar dado por:

$$\begin{bmatrix}
d^2x & dxdy \\
dydx & d^2y 
\end{bmatrix} = \begin{bmatrix}
0 & \frac{-1}{y^2}\\
\frac{-1}{y^2} & \frac{2x}{y^3} 
\end{bmatrix}$$

Si calculamos su determinante tenemos: $\frac{2x}{y^3} - \frac{1}{y^4}$ pero aqui no podemos afirmar que esto siempre es positivo. por lo tanto no es convexa la función. Si hacemos un proceso análogo con $-f$ llegamos a la misma conclusión y por lo tanto no va a ser cóncavo.

Esta función si va a ser quasiconvexa ya que tomando los conjuntos de nivel:

$$\{(x,y) | \frac{x}{y} \leq \alpha\} \Leftrightarrow \{(x,y) | x \leq \alpha y\}$$

Y esto se puede expresar como $\begin{bmatrix}
1 \\
-\alpha 
\end{bmatrix}^T * \begin{bmatrix}
x\\
y 
\end{bmatrix} \leq 0$ 
lo cual sería un semiespacio, por lo tanto es convexo. Para mirar quasiconcavidad es un proceso análogo
<br>
**(e)** Utilicemos el criterio de segundo orden, entonces con:
$$f(x,y) = \frac{x^2}{y} \qquad dx=\frac{2x}{y} \qquad dy=\frac{-x^2}{y^2}$$<br>

Luego su hessiano va a estar dado por:

$$\begin{bmatrix}
d^2x & dxdy \\
dydx & d^2y 
\end{bmatrix} = \begin{bmatrix}
\frac{2}{y} & \frac{-2x}{y^2}\\
\frac{-2x}{y^2} & \frac{2x^2}{y^3} 
\end{bmatrix}$$

La matriz va a ser positiva semidefinida ya que los determinantes de las submatrices principales superiores son positivos o 0 ($\frac{2}{y} > 0$ y $\frac{4x^2}{y^4} - \frac{4x^2}{y^4} = 0$ por lo tanto es convexa y quasiconvexa.

No es concava ya que el hessiano de $-f$

$$\begin{bmatrix}
\frac{-2}{y} & \frac{2x}{y^2}\\
\frac{2x}{y^2} & \frac{-2x^2}{y^3} 
\end{bmatrix}$$

No es positivo semidefinido ya que $\frac{-2}{y} < 0$<br>

**(f)** Utilicemos el criterio de segundo orden para concavas, entonces con $-f$:
$$-f(x,y) = -\frac{x^\alpha}{y^{\alpha-1}} \qquad dx=-\frac{\alpha x^{\alpha-1}}{y^{\alpha-1}} \qquad dy=-\frac{-(\alpha-1)x^\alpha}{y^\alpha}$$

Luego su hessiano va a estar dado por:

$$\begin{bmatrix}
d^2x & dxdy \\
dydx & d^2y 
\end{bmatrix} = \begin{bmatrix}
\frac{-\alpha(\alpha-1)x^{\alpha-2}}{y^{\alpha-1}} & \frac{(\alpha-1)\alpha x^{\alpha-1}}{y^\alpha}\\
\frac{(\alpha-1)\alpha x^{\alpha-1}}{y^\alpha} & \frac{-(\alpha-1)\alpha x^\alpha}{y^{\alpha+1}} 
\end{bmatrix}$$

$$ = -\alpha(1-\alpha)x^{\alpha}y^{\alpha-1}
\begin{bmatrix}
\frac{1}{x^2} & \frac{-1}{xy}\\
\frac{-1}{xy} & \frac{1}{y^2}
\end{bmatrix}
$$

Luego es claro que por las determinantes de las submatrices principales superiores esta matriz va a ser semidefinida positiva, por lo tanto la función es concava luego quasiconcava.

**3.20** *Composition with an affine function*. Show that the following functions $f: R^n \rightarrow R$ are convex.

(a) $f(x) = ||Ax-b||$, where $A \in R^{mxn}, b \in R^m$, and $||\cdot||$ is a norm on $R^m$. <br>
(b) $f(x) = -(det(A_0 + x_1A_1 + ... + x_nA_n))^{1/m}$, on $\{x | A_0 + x_1A_1 + ... + x_nA_n \succ 0 \},$ where $A_i \in S^m$. <br>

**(b)**. Primero veamos que $h(X)=(det(X))^{1/m}$ es concava sobre $dom f=S_{++}^m$:

De (1) vemos que $g$ es una función concava de $t$ sobre $\{t|Y+tV≻0\}$ pues $detY≻0$ y la media geometrica $(∏{i=1}^mx_i)^{1/m}$ es concava sobre $R^n{++}$. 

De modo que $-h(X)$ es una función convexa. Ahora, note que $f$ es la composición de la funcion convexa $-(det(X))^{1/m}$ y una transformación afin, por lo tanto $f$ es convexa.

**3.21** *Pointwise maximum and supremum*. Show that the following functions $f: R^n \rightarrow R$ are convex.

(a) $f(x) = max_{i=1m,...,k} ||A^{(i)}x- b^{(i)}||$, where $A^{(i)} \in R^{m \times n}, b^{(i)} \in R^m$ and $||\cdot||$ is a norm on $R^m$. <br>

In other words, $|x|_{[1]}, |x|_{[2]}, ... ,|x|_{[n]}$ are the absolute values of the components of $x$, sortes in nonincreasing order.

**Solución:**

**(a)**. Note que $||A^{(i)}x-b^{(i)}||$ es la composición de una norma y una transformacion afin , por lo tanto es una función convexa. De modo que $f$ es convexa pues es el máximo punto a punto de las $k$ funciones $||A^{(i)}x-b^{(i)}||$.

**(b)**. Note que $f(x)= \sum_{i=1}^r |x|{[i]} = \max \{x{i_1}+⋯,x_{i_r}|1\leq i_1 < i_2 < ⋯ < i_r\leq n\}$.

**3.25** *Maximum probability distance between distributions*. Let $p, q \in R^n$ represent two probability distributions on $\{1,...,n\}$ (so $p,q \succeq 0, 1^Tp = 1^Tq = 1)$. We define the *maximum probability distance* $d_{mp}(p,q)$ between $p$ and $q$ as the maximum difference in probability assigned by $p$ and $q$, over all events:

$$ d_{mp}(p,q) = max\{|prob(p,C) - prob(q,C)|| C \subseteq \{1,...,n\}\}$$

Here $prob(p,C)$ is the probability of $C$, under the distribution $p$, *i.e.*, $prob(p,C) = \sum_{i\in C}p_i$.

Find a simple expression for $d_{mp}$, involving $||p-q||_1 = \sum_{i=1}^n|p_i - q_i|$, and show that $d_{mp}$ is a convex function on $R^n \times R^n$. (Its domain is $\{(p,q) | p,q \succeq 0, 1^Tp = 1^Tq = 1\}$, but it has a natural extension to all of $R^n \times R^n$).

**Solución:**

Considerando que tenemos distribuciones de probabilidad se tiene que  $prob(p,C)=1-prob(p,\overline{C}) \ y \ prob(q,C)=1-prob(q,\overline{C}) $, donde $\overline{C}$ es el complemento de $C$, de esta manera se tiene que 

$$ prob(p,C)-prob(q,C)=1-prob(p,\overline{C})-1+prob(q,\overline{C})=-prob(p,\overline{C})+prob(q,\overline{C})$$

con lo cual simplificamos la expresión 

$$d_{mp}(p,q)=max\{prob(p,C)-prob(q,C)|C\subseteq \{1,\dots ,n\}\}$$

considerando la función $f_C(p,q)=prob(p,C)-prob(q,C)$ se tiene entonces que $d_{mp}$ es el máximo de un conjunto de funciones $\{f_C(p,q)\}_{C\subseteq \{1,\dots n\}}$ las cuales son afines, es decir convexas, por lo tanto $d_{mp}$ es convexa.

Para simplificar la expresión queremos buscar un $C$ que maximiza $prob(p,C)-prob(q,C)=\sum_{C}p_i-q_i$ , veamos que este $C$  tiene la forma $C=\{i\in \{1,\dots n\}|p_i>q_i\}$. 

Consideremos otro $C^*\subseteq\{1,\dots n\}$ de tal manera que existe un $k\in C$ y $k\notin C^*$, (no consideramos el caso $p_i=q_i$ pues estos no le aportan a la sumatoria), luego adicionando este $k$ a $C^*$, se tendría que $prob(p,C^*)-prob(q,C^*)$ incrementa en $p_k-q_k>0$, por lo que $C$ no maximiza la diferencia, suponiendo ahora un $k\notin C$ y $k\in C^*$ luego $p_k<q_k$ por lo que si quitamos $k$ en $C^*$, se tendría un incremento  en $prob(p,C^*)-prob(q,C^*)$ por lo que tampoco podría ser óptimo, de esta manera podemos expresar 

$$d_{mp}=\sum_{p_i>q_i}(p_i-q_i)$$

por otro lado se tiene que $\sum_{i=1}^n(p_i-q_i)=0$ pues $p,q$ son distribuciones de probabilidad, luego separando esta sumatoria, la podemos expresar como

$$\sum_{p_i>q_i}(p_i-q_i)+\sum_{p_i\leq q_i}(p_i-q_i)=0$$

con lo que se tiene que $d_{mp}=\sum_{p_i>q_i}(p_i-q_i)=-\sum_{p_i\leq q_i}(p_i-q_i)$ es decir 

$$
    d_{mp}=\frac{1}{2}(\sum_{p_i>q_i}(p_i-q_i)-\sum_{p_i\leq q_i}(p_i-q_i))
$$

luego se tiene que 

$$
    d_{mp}=\frac{1}{2}(\sum_{i=1}^n|(p-q)|)=\frac{\|p-q\|_1}{2}
$$

**3.29** *Representation of piecewise-linear convex functions.* A convex function $f: R^n \rightarrow R$, with $dom f = R^n$, is called *piecewise-linear* if there exists a partition of $R^n$ as

$$R^n = X_1 \cup X_2 \cup ... \cup X_L$$,

where $int X_i \neq \emptyset$ and $int X_i \cap int X_j = \emptyset$ for $i \neq j$, and a family of affine functions $a_1^Tx + b_1, ... , a_L^Tx + b_L$ such that $f(x) = a_i^Tx + b_i$ for $x\in X_i$.

Show that such a function has the form $f(x) = max \{a_1^Tx + b_1, ... , a_L^Tx + b_L\}$.

**Solución**:

Por la desigualdad de Jensen sabemos que $∀x,y \in domf$ y $0 \leq \theta \leq 1$ se tiene

$f(\theta x + (1- \theta )y)=f(y+ \theta(x-y)) \leq f(y) + \theta (f(x)-f(y)) $. Así

$(f(y+\theta(x-y))-f(y))(1/\theta) \leq f(x) - f(y)$ (1)

Ahora, supongamos que $x \in X_i$ y escojamos cualquier $y \in intX_j$, para algún $j \not= i$, y al tomar un $\theta$ lo suficientemente pequeño de forma tal que se satisfaga $y+\theta(x-y) \in X_j$ obtenemos que (1) se puede reescribir como

$a_j^Ty+b_j+\dfrac{a_j^T(y+\theta(x-y))+b_j-a_j^Ty-b_j}{\theta}=a_j^Tx+b_j\leq a_i^T+b_i$

Como el $j$ fue elegido arbitrariamente, $\max_{j=1,\cdots,L}(a_j^Tx+b_j)\leq a_i^Tx+b_i$. Por lo tanto, $a_i^Tx+b_i=\max_{j=1,\cdots,L}(a_j^Tx+b_j)$.

**3.32** *Products and ratios of convex functions.* In general the product or ratio of two convex functions is not convex. However, there are some results that apply to functions on $R$. Prove the following

(a) If $f$ and $g$ are convex, both nondecreasing (or nonincreasing), and positive functions on an interva, then $fg$ is convex.<br>
(b) If $f,g$ are concave, positive, with one nondecreasing and the other nonincreasing, then $fg$ is concave.<br>
(c) If $f$ is convex, nondecreasing, and positive, and $g$ is concave, nonincreasing and positive, then $f/g$ is convex.

**Solución**:

**(a)**

como $f,g $ son convexas entonces para un $0\leq\theta\leq 1$ y $x,y \in dom(fg)$ se tiene 

$$\begin{aligned}
f(\theta x +(1-\theta)y)\leq \theta f(x)+(1-\theta)f(y) \\
g(\theta x +(1-\theta)y)\leq \theta g(x)+(1-\theta)g(y)
\end{aligned}$$


y como $f,g $ son positivas entonces 

$$\begin{aligned}
f(\theta x +(1-\theta)y)g(\theta x +(1-\theta)y)&\leq \theta f(x)g(x)+(1-\theta)f(y)g(y) \\
&=\theta f(x)g(y)+(1-\theta)f(y)g(y)+\theta(1-\theta)(f(y)-f(x))(g(x)-g(y))\\
&\leq \theta f(x)g(y)+(1-\theta)f(y)g(y)
\end{aligned}$$

esta última desigualdad se obtiene apartir de que $f,g$ no son decrecientes pues el término $\theta(1-\theta)(f(y)-f(x))(g(x)-g(y))$ sería cero o negativo. De esta manera se tiene que $fg$ es una función convexa.

**(b)**
como $f,g$ son concavas entonces $-f,-g$ son convexas y realizamos el producto y llegamos a la desigualdad del anterior punto pero invertida 

$$\begin{aligned}
f(\theta x +(1-\theta)y)g(\theta x +(1-\theta)y)&\geq \theta f(x)g(x)+(1-\theta)f(y)g(y) \\
&=\theta f(x)g(y)+(1-\theta)f(y)g(y)+\theta(1-\theta)(f(y)-f(x))(g(x)-g(y))\\
&\geq \theta f(x)g(y)+(1-\theta)f(y)g(y)
\end{aligned}$$

y lo último se tiene porque una función no es creciente y la otra no decreciente, por lo que el término $\theta(1-\theta)(f(y)-f(x))(g(x)-g(y))$ es positivo, y finalmente se invierte la desigualdad y se llega a que $-fg$ es convexo, luego $fg$ es concava.

**(c)**
Observemos que podemos ver $\frac{1}{g}=exp(-log(g))$ es decir es una función convexa y como $g$ es no creciente entonces $\frac{1}{g}$ es no decreciente,de esta manera se tendría que $\frac{f}{g}$ es una función convexa, pues tendríamos el caso del punto **(a)**

is log-concave. This follows from the general result that the convolution of two log-concave functions ins log-concave. In this problem we guide you through a simple self-contained proof that $f$ is log-concave if and only if $f''(x) f(x) \leq f'(x)^2$ for all $x$.

(a) Verify that $f''(x) f(x) \leq f'(x)^2$ for $x \geq 0$. That leaves us the hard part, which is to show the inequality for $x < 0$. <br>
(b) Verify that for any $t$ and $x$ we have $t^2/2 \geq -x^2/2 + xt$. <br>
(c) Using part (b) show that $e^{-t^2/2} \leq e^{x^2/2 - xt}$. Conclude that, for $x < 0$,

$$\int_{-\infty}^x e^{-t^2/2}dt \leq e^{x^2/2}\int_{\infty}^x e^{-xt} dt$$

(d) Use part (c) to verify that $f''(x)f(x) \leq f'(x)^2 $ for $x \leq 0$

**Solución:**

**(a)**
tomando las derivadas de $f(x)$ se tiene

$$\begin{aligned}
f'(x)=\frac{e^{-\frac{x^2}{2}}}{\sqrt{2\pi}}\\
f''(x)=\frac{-xe^{-\frac{x^2}{2}}}{\sqrt{2\pi}}
\end{aligned}$$


luego como $x\geq 0$ entonces $f''(x)<=0$ y además $f(x)\geq 0$ luego $f''(x)f(x)\leq 0$ entonces 

$$f''(x)f(x)\leq f'(x)^2$$

**(b)**
Se tiene $0\leq (x-t)^2=x^2-2xt+t^2$ reorganizando la desigualdad se tiene que 

$-x^2+2xt\leq t^2$ es decir $-\frac{x^2}{2}+xt\leq \frac{t^2}{2}$

**(c)**

Invirtiendo la desigualdad anterior y como $e^x$ es una función creciente se tiene que 
$$e^{-\frac{t^2}{2}}\leq e^{\frac{x^2}{2}-xt}$$
como $e^x$ es además una función no negativa, entonces podemos integrar a ambos lados de la desigualdad y se obtiene
$$\int_{-\infty}^x e^{-\frac{t^2}{2}}dt \leq \int_{-\infty}^x  e^{\frac{x^2}{2}-xt}dt= e^{\frac{x^2}{2}}\int_{-\infty}^x  e^{-xt}dt$$

**(d)**
De la última desigualdad podemos desarrollar la integral de la derecha y tendríamos


$$\begin{aligned}
\int_{-\infty}^x e^{-\frac{t^2}{2}}dt &\leq  e^{\frac{x^2}{2}}\int_{-\infty}^x  e^{-xt}dt \\
&=e^{\frac{x^2}{2}} (-\frac{1}{x})(e^{-x^2})
\end{aligned}$$

como $x<0$ entonces reorganizando la desigualdad llegamos a que 
$$-xe^{\frac{-x^2}{2}}\int_{-\infty}^x e^{-\frac{t^2}{2}}dt\leq e^{-x^2}$$
tomando $\alpha=\frac{1}{\sqrt(2\pi)}$ y multiplicando por $\alpha^2$ a cada lado y asociando, se obtiene
$$-\alpha xe^{\frac{-x^2}{2}}\alpha\int_{-\infty}^x e^{-\frac{t^2}{2}}dt\leq \alpha^2 e^{-x^2}$$

se tiene entonces 
$$f''(x)f(x)\leq f'(x)^2$$
